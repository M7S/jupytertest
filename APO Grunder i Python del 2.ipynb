# Upprepning med for-loop
Vill man att programmet ska skriva ut de 100 första heltalen kan man förstås skriva 100 st print-kommandon. Men det blir ganska jobbigt i längden så i alla programmeringsspråk finns det kommandon för att upprepa saker många gånger. I python används ofta s.k. for-loopar. Testa koden och förklara vad den gör.

```python
for tal in range(100):                      # Skriv förklaring till vad koden gör här.
  print(tal)                                # Skriv förklaring till vad koden gör här.
print("Denna print-kod körs bara en gång")  # Körs bara en gång, eftersom den är utanför loopen, inget mellanslag först.           
```

I följande program beräknar vi summan av alla heltal från 0 till 100.

```python
summa = 0                 # Inför en variabel som heter summa och som vi först ger värdet 0
for tal in range(101):    # Inför en variabel som heter tal och antar värdena 0,1,2,3,...,100
  summa = summa + tal     # Varje gång loopen körs uppdateras summan genom att addera till tal
  print(summa)            # Skriver ut summan, varje steg skilt
```
