**Du behöver inte förstå den här koden, bara använda den!**

````
# Installera ColabTurtle
!pip3 install ColabTurtle

# Importera funktionerna
from ColabTurtle.Turtle import *

# Förbättra goto-funktionen
ogoto = goto

def goto(x,y):
  hw = window_size[0] // 2
  hh = window_size[1] // 2
  if not isinstance(x, int):
        raise ValueError('New x position should be an integer')
  if x < -hw or x > hw:
        raise ValueError('New x position outside the screen.')
  if not isinstance(y, int):
        raise ValueError('New y position should be an integer')
  if y < -hh or y > hh:
        raise ValueError('New y position is outside the screen')
  ogoto(x + hw, -y + hh)

# En funktion som startar ColabTurtle med koordinataxlar utritade.
def initializeTurtleWithAxis(initial_speed=DEFAULT_SPEED, initial_window_size=DEFAULT_WINDOW_SIZE):
  initializeTurtle(initial_speed, initial_window_size)
  # rita koordinataxlarna
  bgcolor("white")
  color("black")
  width(1)
  speed(10)
  penup()
  hw = window_size[0] // 2
  hh = window_size[1] // 2
  goto(-hw,0)
  pendown()
  goto(hw,0)
  goto(hw-5,-4)
  penup()
  goto(hw-5,4)
  pendown()
  goto(hw,0)
  penup()
  goto(0, hh-1)
  pendown()
  goto(4, hh-6)
  penup()
  goto(-4, hh-6)
  pendown()
  goto(0, hh-1)
  goto(0, -hh)
  # Återgå till startinställningarna
  penup()
  goto(0,0)
  pendown()
  color("red")
  width(DEFAULT_PEN_WIDTH)
  speed(initial_speed)
````

````
ålder = input("Hur gammal är du?")
ålder = int(ålder)
if ålder >= 15:
  print("Du har uppnått mopedåldern.")
else:                                                             # Skriv förklaring till vad koden gör här.
  print("Du är inte tillräckligt gammal för att köra moped än.")
print("Roligt att chatta med dig, hej då.")
````

Det finns ett antal olika jämförelser du kan använda i python.

```
==  lika med (Obs! Två likamedtecken efter varandra!)
<   mindre än
>   större än
<=  mindre än eller lika med
>=  större än eller lika med
!=  inte lika med
```
