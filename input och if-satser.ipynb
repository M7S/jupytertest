Ofta önskaar man att ens program ska göra olika saker beroende på vad användaren svarar på någon fråga.

Vad gör följande program? Kör programmet och kolla om du hade rätt.

```python
# Skapar en variabel ålder. Men för att kunna använda svaret som ett talvärde 
# måste vi berätta åt datorn att det är ett heltal (int) eller ett decimaltal (float)

ålder = int(input("Hur gammal är du? ")) 
if ålder < 18:    
  print("Du är inte ännu myndig.") 
else:
  print("Du är myndig.")
```

Vad gör följande program?

```python
poäng = int(input("Hur många poäng fick du i förra matematikprovet? "))

if poäng >= 29:
  vitsord = 10
elif poäng >= 25:
  vitsord = 9
elif poäng >= 21:
  vitsord = 8
elif poäng >= 17:
  vitsord = 7
elif poäng >= 13:
  vitsord = 6
elif poäng >=9:
  vitsord = 5
else:
  vitsord = 4

print("Du fick vitsordet:",vitsord,)
```
