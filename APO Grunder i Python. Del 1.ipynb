# input()
input() är en funktion som låter användaren skriva något som sedan **returneras** så att det kan sparas i en variabel.

Läs igenom koden här under försök lista ut vad koden gör. Testa sedan koden för att se om du har gissat rätt. **Skriv sedan in vad koden gör bakom #-tecknen.**

````
namn = input("Hej, vad heter du? ")   # Skriv förklaring till vad koden gör här.
print(namn, "är ett fint namn!")      # Skriv förklaring till vad koden gör här.
````

För att det som användaren skriver ska sparas måste man alltså börja med berätta i vilken variabel texten ska sparas.

# Olika typer av variabler

Här nedanför är ett annat exempel. Vad tror du att programmet kommer att skriva på skärmen? Testa för att se om du har rätt. Skriv sedan in förklaringar.

````
tal = input("Skriv ett tal. ")      # Skriv förklaring till vad koden gör här.
produkt = tal * 3                   # Skriv förklaring till vad koden gör här.
print(tal, "* 3 =", produkt)        # Skriv förklaring till vad koden gör här.
````

Blev resultatet annorlunda än du hade förväntat dig? Det beror på att input alltid ger en text. Det som sparades i variabeln tal var alltså en text även om det var en text som bestod av siffror. Tar du en text gånger tre får du samma text tre gånger efter varandra. Exemplet nedanför visar det här.

````
print(7 * 5)        # Skriv förklaring till vad koden gör här.
print("7" * 5)      # Skriv förklaring till vad koden gör här.
````

Så hur ska vi då göra om vi vill räkna matematik med något som kommer från input()? I python finns det en funktion som heter int() som gör om text (och annat) till heltal (integer). Läs igenom programmet nedanför och testa det sedan. 

````
tal = input("Skriv ett tal. ")      # Skriv förklaring till vad koden gör här.
tal = int(tal)
produkt = tal * 3                   # Skriv förklaring till vad koden gör här.
print(tal, "* 3 =", produkt)        # Skriv förklaring till vad koden gör här.
````
