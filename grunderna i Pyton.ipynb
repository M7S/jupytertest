# input()
input() är en funktion som låter användaren skriva något som sedan **returneras** så att det kan sparas i en variabel.

Läs igenom koden här under försök lista ut vad koden gör. Testa sedan koden för att se om du har gissat rätt.

````
namn = input("Hej, vad heter du? ")
print(namn, "är ett fint namn!")
````

För att det som användaren skriver ska sparas måste man alltså börja med berätta i vilken variabel texten ska sparas.

# Olika typer av variabler

Här nedanför är ett annat exempel. Vad tror du att programmet kommer att skriva på skärmen? Testa för att se om du har rätt.

````
tal = input("Skriv ett tal. ")
produkt = tal * 3
print(tal, "* 3 =", produkt)
````

Blev resultatet annorlunda än du hade förväntat dig? Det beror på att input alltid ger en text. Det som sparades i variabeln tal var alltså en text även om det var en text som bestod av siffror. Tar du en text gånger tre får du samma text tre gånger efter varandra. Exempelt nedanför visar det här.

````
print(7 * 5)
print("7" * 5)
````

Så hur ska vi då göra om vi vill räkna matematik med något som kommer från input()? I python finns det en funktion som heter int() som gör om text (och annat) till heltal. Läs igenom programmet nedanför och testa det sedan. 

````
tal = input("Skriv ett tal. ")
tal = int(tal)
produkt = tal * 3
print(tal, "* 3 =", produkt)
````
